# Upload Images

```python
!mkdir -p /content/gaussian-splatting/data/input
!cd /content/gaussian-splatting

from google.colab import files
uploaded = files.upload()
```

```python
import os
import shutil
from PIL import Image

# Check if a file is a valid image image or not
def is_image(file_path):
    try:
        # Open the file using PIL
        with Image.open(file_path) as img:
            return True
    except:
        return False

# List the uploaded files
uploaded_files = list(uploaded.keys())

# Move the uploaded files to the "data/input" folder
for file_name in uploaded_files:
    source_path = file_name
    if is_image(source_path):
      destination_path = os.path.join("/content/gaussian-splatting/data/input", file_name)
      shutil.move(source_path, destination_path)
```
